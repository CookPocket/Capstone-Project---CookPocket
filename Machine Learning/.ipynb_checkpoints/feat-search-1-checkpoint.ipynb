```python
import pandas as pd

# Load dataset
file_path = './Data/all_data_resep.csv'
data = pd.read_csv(file_path)

# Menampilkan beberapa baris pertama untuk memahami struktur dataset
data.head()
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download('wordnet')
nltk.download('stopwords')

# Gabungkan kolom-kolom yang relevan menjadi representasi teks tunggal
data['Combined_Text'] = data['Title'] + " " + data['Ingredients'] + " " + data['Steps']

# Tambahkan threshold untuk kolom 'Loves'
threshold = 100
data['Label'] = (data['Loves'] >= threshold).astype(int)

# Bersihkan teks: huruf kecil, hapus karakter khusus, dan hapus spasi
def clean_text(text):
    # Konversi ke huruf kecil
    text = text.lower()
    
    # Menghapus tanda baca dan karakter non-huruf
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    
    # Tokenisasi (memisah kata-kata)
    words = text.split()
    
    # Hapus stopwords
    stop_words = set(stopwords.words('indonesian'))  
    words = [word for word in words if word not in stop_words]
    
    # Stemming atau Lemmatization
    lemmatizer = WordNetLemmatizer()
    words = [lemmatizer.lemmatize(word) for word in words]
    
    return ' '.join(words)

data['Cleaned_Text'] = data['Combined_Text'].apply(clean_text)

# Menampilkan dataset yang telah dibersihkan dengan kolom baru
data[['Title', 'Combined_Text', 'Cleaned_Text']].head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory

# Inisialisasi TF-IDF Vectorizer
factory = StopWordRemoverFactory()
stopword_remover = factory.create_stop_word_remover()
data['Cleaned_Text'] = data['Cleaned_Text'].apply(stopword_remover.remove)

# Access the internal stop words list (this might be implementation-specific)
sastrawi_stop_words = factory.get_stop_words() 
tfidf_vectorizer = TfidfVectorizer(max_features=500, stop_words=sastrawi_stop_words)  

# Fit dan transform the cleaned text
tfidf_matrix = tfidf_vectorizer.fit_transform(data['Cleaned_Text'])

# Ubah matriks TF-IDF menjadi DataFrame agaroembacaan yang lebih baik
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf_vectorizer.get_feature_names_out())

tfidf_df.head()
```

```python
# Shuffle Data
dataset_shuffled = data.sample(frac=1, random_state=42).reset_index(drop=True)

# Check the first few rows after shuffling
dataset_shuffled.head()
```

```python
from gensim.models import Word2Vec
import numpy as np

# Tokenisasi data teks yang sudah dibersihkan
data['Tokenized_Text'] = data['Cleaned_Text'].apply(lambda x: x.split())

# Melatih model Word2Vec
word2vec_model = Word2Vec(
    sentences=data['Tokenized_Text'], 
    vector_size=100,  # Dimensi vektor embedding
    window=5,  # Jangkauan konteks
    min_count=1,  # Minimal frekuensi kata
    workers=4,  # Jumlah thread
    sg=1  # Gunakan Skip-Gram (sg=1) atau CBOW (sg=0)
)

# Simpan model untuk digunakan kembali
word2vec_model.save("word2vec_resep.model")
```

```python
# Fungsi untuk menghitung rata-rata embedding sebuah dokumen
def document_embedding(text, model, vector_size=100):
    words = text.split()
    word_vectors = [model.wv[word] for word in words if word in model.wv]
    if len(word_vectors) > 0:
        return np.mean(word_vectors, axis=0)  # Rata-rata embedding
    else:
        return np.zeros(vector_size)  # Jika kata tidak ditemukan, gunakan nol

# Terapkan fungsi ini ke seluruh dataset
data['Embedding'] = data['Cleaned_Text'].apply(
    lambda x: document_embedding(x, word2vec_model, vector_size=100)
)

# Ubah kolom embedding menjadi array NumPy
embeddings = np.array(data['Embedding'].tolist())
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras import layers, models
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout, Input

# Tentukan bentuk input berdasarkan fitur TF-IDF
# input_shape = tfidf_matrix.shape[1]

# Membangun model
# model = models.Sequential([
#     layers.InputLayer(input_shape=(embeddings.shape[1],)),
#     layers.Dense(128, activation='relu'),
#     layers.BatchNormalization(),
#     layers.Dropout(0.3),  
#     layers.Dense(64, activation='relu'),
#     layers.BatchNormalization(),
#     layers.Dropout(0.3),
#     layers.Dense(32, activation='relu'),
#     layers.BatchNormalization(),
#     layers.Dropout(0.3),
#     layers.Dense(1, activation='sigmoid') 
# ])

# Parameter
vocab_size = 10000  # Sesuai jumlah kata di Tokenizer
embedding_dim = 16  # Dimensi embedding
max_length = 100

# Definisi model
model = Sequential([
    Input(shape=(max_length,)),
    Embedding(vocab_size, embedding_dim),
    LSTM(32, return_sequences=False),
    Dropout(0.2),
    Dense(16, activation='relu'),
    Dropout(0.2),
    Dense(1, activation='sigmoid')  # Karena klasifikasi biner
])

# Compile model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.summary()
```

```python
import numpy as np

labels = np.random.randint(0, 2, size=(tfidf_matrix.shape[0],))

# Membagi data menjadi set pelatihan dan validasi
from sklearn.model_selection import train_test_split
# X_train, X_val, y_train, y_val = train_test_split(tfidf_matrix.toarray(), labels, test_size=0.2, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(embeddings, labels, test_size=0.2, random_state=42)
```

```python
print(X_train.shape)
print("Jumlah data training:", len(X_train))
print("Jumlah data validation:", len(X_val))
```

```python
early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor='val_accuracy',  
    # patience=5,               
    min_delta=0.01,           
    mode='max',               
    verbose=1,                
    restore_best_weights=True
)
```

```python
# Train the model
history = model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=100,
    batch_size=32,
    # callbacks=[early_stopping]
)
```

```python
# Evaluasi model
loss, accuracy = model.evaluate(X_val, y_val)
print(f"Validation Loss: {loss}")
print(f"Validation Accuracy: {accuracy}")
```

```python
import matplotlib.pyplot as plt

# Plot loss
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.legend()
plt.title("Loss")
plt.show()

# Plot accuracy
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.legend()
plt.title("Accuracy")
plt.show()
```
